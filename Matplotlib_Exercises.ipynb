## Exercise 2
** Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

```python
fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])

ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])

plt.show()
```

** Now plot (x,y) on both axes. And call your figure object to show it.**

```python
fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])

ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])

x = np.arange(0,100)
y = x*2

ax1.plot(x, y, 'r')  
ax1.set_xlabel('X-axis Label for ax1')
ax1.set_ylabel('Y-axis Label for ax1')
ax1.set_title('Title for ax1')

ax2.plot(x, y, 'r')  
ax2.set_xlabel('X-axis Label for ax2')
ax2.set_ylabel('Y-axis Label for ax2')
ax2.set_title('Title for ax2')

plt.show()
```

## Exercise 3

** Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
fig = plt.figure()

# Add the first axis located at [0, 0, 1, 1]
ax1 = fig.add_axes([0, 0, 1, 1])

# Add the second axis located at [0.2, 0.5, 0.4, 0.4]
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

plt.show()
```

** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])

ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

x = np.arange(0,100)
y = x*2
z=x**2

ax1.plot(x, z, 'b')  
ax1.set_xlabel('x')
ax1.set_ylabel('z')

ax2.plot(x, y, 'r')  
ax2.set_xlabel('x')
ax2.set_ylabel('y')

plt.show()
```

## Exercise 4

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

axes[0].plot(x, y, color='blue', linestyle=':', linewidth=2, label='(x, y)')
axes[0].set_xlabel('X-axis Label for Left Plot')
axes[0].set_ylabel('Y-axis Label for Left Plot')
axes[0].set_title('Left Plot')
axes[0].legend()

axes[1].plot(x, z, color='red', linestyle='-', linewidth=3, label='(x, z)')
axes[1].set_xlabel('X-axis Label for Right Plot')
axes[1].set_ylabel('Z-axis Label for Right Plot')
axes[1].set_title('Right Plot')
axes[1].legend()

plt.tight_layout()

plt.show()
```

** See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,3))  # Adjust figsize as needed

axes[0].plot(x, y, color='blue', linestyle=':',linewidth=2, label='(x, y)')
axes[0].set_xlabel('X-axis')
axes[0].set_ylabel('Y-axis')
axes[0].legend()

axes[1].plot(x, z, color='red', linestyle='-', linewidth=3, label='(x, z)')
axes[1].set_xlabel('X-axis')
axes[1].set_ylabel('Z-axis')
axes[1].legend()

plt.tight_layout()

plt.show()
```
